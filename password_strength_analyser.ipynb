```python
import re

class PasswordStrengthAnalyzer:
    def __init__(self, dictionary=None):
        # Accept an in-memory dictionary for dictionary-based attacks
        self.dictionary = set(dictionary) if dictionary else set()

    def analyze(self, password):
        """Assess the strength of a password."""
        score = 0
        issues = []

        # 1. Length check
        if len(password) >= 12:
            score += 2
        elif len(password) >= 8:
            score += 1
        else:
            issues.append("Password is too short.")

        # 2. Character variety check
        if re.search(r'[A-Z]', password):
            score += 1
        else:
            issues.append("Add uppercase letters for better strength.")

        if re.search(r'[a-z]', password):
            score += 1
        else:
            issues.append("Add lowercase letters for better strength.")

        if re.search(r'[0-9]', password):
            score += 1
        else:
            issues.append("Add digits for better strength.")

        if re.search(r'[!@#$%^&*(),.?":{}|<>]', password):
            score += 1
        else:
            issues.append("Add special characters for better strength.")

        # 3. Dictionary-based check
        if password.lower() in self.dictionary:
            issues.append("Password is a common word found in the dictionary.")

        # Strength rating
        if score >= 6:
            rating = "Strong"
        elif score >= 4:
            rating = "Moderate"
        else:
            rating = "Weak"

        return {
            "rating": rating,
            "score": score,
            "issues": issues
        }

def main():
    # Provide an example dictionary for testing
    example_dictionary = [
        "password", "123456", "qwerty", "letmein", "welcome", "iloveyou"
    ]

    analyzer = PasswordStrengthAnalyzer(dictionary=example_dictionary)

    while True:
        print("\nEnter a password to analyze (or 'exit' to quit):")
        password = input("Password: ").strip()
        if password.lower() == "exit":
            print("Exiting Password Strength Analyzer. Goodbye!")
            break

        result = analyzer.analyze(password)
        print(f"\nPassword Strength: {result['rating']}")
        print(f"Score: {result['score']}/6")
        if result['issues']:
            print("Recommendations to improve your password:")
            for issue in result['issues']:
                print(f"  - {issue}")
        else:
            print("Your password meets all the strength criteria!")

# Run the main function
main()
```
